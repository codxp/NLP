```python
import re
```

```python
# Define a simple pattern

pattern = r'\d+' # Example: matches one or more digits

# Create a regex object using a re.compile()
regex_object = re.compile(pattern)
```

```python
text = 'There are 123 mangos and 456 bananas'

# Using a re.search() with a pre-compiled regex object
match_text = regex_object.search(text)
if match_text:
  print(f'Pattern found: {match_text.group()}')
else:
  print('Pattern not found')
```

```python
# Patterns to extract multiple address in a text

pattern = r'\d+[A-Za-z]+'

text = 'This is my address: Apple Street 36aB and This is my address: Apple Street 756AJ'

regex_obj = re.compile(pattern)
match_text = regex_obj.findall(text)

if match_text:
  print(f'Pattern found: {match_text}')
else:
  print('Pattern not found')
```

```python
# Extract e-mail with regex:

text = 'This is my personnal e-mail alice@google.com and this is my office e-mail alice22@abc.com'

email_pattern = r'(\w+@[a-zA-Z_]+?\.[a-zA-Z]{2,6})'
regex_obj = re.compile(email_pattern)

match_text = regex_obj.findall(text)

if match_text:
  print(f'Pattern found: {match_text}')
else:
  print('Pattern not found')
```

```python
# Extract phone number with regex:

text = 'the following are the phone numbers of three students in a class, Michael : +234-123-456-7890, James : +234 100 000 1000, Sarah : +234999999999'

phone_pattern = r'\+[0-9\s\_]+'

regex_obj = re.compile(phone_pattern)

match_text = regex_obj.findall(text)

if match_text:
  print(f'Pattern found: {match_text}')
else:
  print('Pattern not found')
```
