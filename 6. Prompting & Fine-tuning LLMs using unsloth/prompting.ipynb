```python
from langchain_openai import AzureChatOpenAI
import os
from dotenv import load_dotenv
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import ChatPromptTemplate
```

```python
load_dotenv()

AZURE_OPENAI_API_KEY = os.getenv('AZURE_OPENAI_API_KEY')
AZURE_OPENAI_ENDPOINT = os.getenv('AZURE_OPENAI_ENDPOINT')
OPENAI_API_VERSION = os.getenv('OPENAI_API_VERSION')
```

```python
llm = AzureChatOpenAI(
    openai_api_version='2024-02-15-preview',
    azure_deployment='flo-gpt4',
    temperature=0)
```

```python
prompt_template = """{question}"""
prompt = ChatPromptTemplate.from_template(prompt_template)
```

```python
# Chain
rag_chain = (
    {"context": str, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
xbox_price_questions = [
    "What is the current price of the Xbox Series X?",
    "How much does the Xbox Series S cost?",
    "Have there been any recent discounts or promotions on Xbox consoles?"
]
```

```python
for question in xbox_price_questions:
    print(rag_chain.invoke({"context": "", "question": question}))
```

```python
from unittest import result


RED = "\033[31m"
YELLOW = "\033[33m"
GREEN = "\033[32m"
RESET = "\033[0m"

# Liste des prompts à tester
prompts = [
    "Provide the following pricing information:",
    "Answer these questions about the Xbox prices:",
    "Give details on the current pricing for Xbox consoles:",
    "Help me understand the latest Xbox prices:",
    "What are the current prices for these Xbox models:",
]

# Contextes différents basés sur le principe AUTOMAT
contexts = [
    "Hey kiddo! Want to know how much the new Xbox consoles cost? Here's what you need to know:",
    "As a hardcore gamer, staying updated on console prices is crucial. Let’s check the latest prices for the Xbox Series X and Series S:",
    "Considering buying a new Xbox? Here’s the latest pricing information you need to make an informed decision:",
    "To help you compare and decide, here's the current pricing for the Xbox Series X and Series S, plus any ongoing promotions:",
    "Curious about how much you’ll need to save for the latest Xbox? Let’s dive into the numbers and maybe find a sweet deal along the way!",
    "For those seriously considering purchasing a new Xbox, here are the most up-to-date prices and any available discounts:",
    "Please provide a comprehensive breakdown of the current prices for the Xbox Series X and Series S, including any recent promotions:",
    "Quickly, what are the prices for the Xbox Series X and Series S right now?",
    "In the current market, it’s important to know the value of gaming consoles. Here’s the latest pricing information for the Xbox Series X and Series S:",
    "You need this information ASAP: What are the current prices of the Xbox Series X and Series S, and are there any recent promotions?"
]

# Description de la tâche (questions sur les prix de la Xbox) et sorties attendues
xbox_price_questions = [
    "What is the current price of the Xbox Series X?",
    "How much does the Xbox Series S cost?",
    "Have there been any recent discounts or promotions on Xbox consoles?"
]

results = []

for context in contexts:
    for prompt in prompts:
        
        prompt_template = prompt + "\n\n" + context + "\n\n" + "\n".join(xbox_price_questions)
        print(f"{YELLOW}{prompt_template}{RESET}")
        response = llm.invoke(prompt_template)
        results.append(response)
        print(f"{GREEN}{response}{RESET}")
        print(f"{RED}{'-'*80}{RESET}")
        
```
