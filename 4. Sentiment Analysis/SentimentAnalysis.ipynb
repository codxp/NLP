### TEXTBLOB:

```python
# Get sentiment score for each review with TEXTBLOB
imbd_review['scores_TextBlob_Pattern'] = imbd_review['review'].apply(lambda s: TextBlob(s).sentiment.polarity)
# Predict sentiment label for each review
imbd_review['pred_TextBlob_Pattern'] = imbd_review['scores_TextBlob_Pattern'].apply(lambda x: 1 if x >=0 else 0)
```

### TEXTBLOB NAIVES BAYES:

```python
from textblob.sentiments import NaiveBayesAnalyzer

imbd_review['pred_TextBlob_NaiveBayes_class'] = imbd_review['review'].apply(lambda s: TextBlob(s, analyzer=NaiveBayesAnalyzer()).sentiment.classification)
imbd_review['scores_TextBlob_NaiveBayes_Pos'] = imbd_review['review'].apply(lambda s: TextBlob(s, analyzer=NaiveBayesAnalyzer()).sentiment.p_pos)
imbd_review['scores_TextBlob_NaiveBayes_Neg'] = imbd_review['review'].apply(lambda s: TextBlob(s, analyzer=NaiveBayesAnalyzer()).sentiment.n_pos)
imbd_review['pred_TextBlob_NaiveBayes'] = imbd_review['pred_TextBlob_NaiveBayes_class'].apply(lambda x: 1 if x == 'pos' else 0)
```

```python
# Take a peek on VADER's lexicon
import nltk

# Download VADER lexicon
nltk.download('vader_lexicon')

# Import the SentimentIntensityAnalyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Create an instance of SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Accessing the lexicon (a dictionary of words and their sentiment scores)
vader_lexicon = sid.lexicon
```

```python
# Get sentiment score for each review
vader_sentiment = SentimentIntensityAnalyzer()
imbd_review['scores_VADER'] = imbd_review['review'].apply(lambda s: vader_sentiment.polarity_scores(s)['compound'])

# Predict sentiment label for each review
imbd_review['pred_VADER'] = imbd_review['scores_VADER'].apply(lambda x: 1 if x >=0 else 0)
```

```python
# Define a function to get Flair sentiment prediction score
def score_flair(text):
  sentence = Sentence(text)
  classifier.predict(sentence)
  score = sentence.labels[0].score
  value = sentence.labels[0].value
  return score, value
```

```python
# Get sentiment score for each review
imbd_review['scores_flair'] = imbd_review['review'].apply(lambda s: score_flair(s)[0])

# Predict sentiment label for each review
imbd_review['pred_flair'] = imbd_review['review'].apply(lambda s: score_flair(s)[1])
```

```python
# Change the label of flair prediction to 0 if negative and 1 if positive
mapping = {'NEGATIVE': 0, 'POSITIVE': 1}
imbd_review['pred_flair'] = imbd_review['pred_flair'].map(mapping)
```
